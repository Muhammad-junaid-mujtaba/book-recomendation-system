```python
import pandas as pd
book =pd.read_csv("/content/best-selling-books.csv")
book
```

```python
book=book[['Book','Author(s)','Original language','Genre']]
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential,Model
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout,Input
from sklearn.preprocessing import LabelEncoder
```

```python
x_train,x_test,y_train,y_test=train_test_split(book['Book'],book['Genre'],test_size=0.2)
```

```python
le=LabelEncoder()

y_train=le.fit_transform(y_train)
y_test=le.fit_transform(y_test)
tokenizer = Tokenizer(num_words=10000) # Example: Consider top 10000 words
tokenizer.fit_on_texts(x_train)

# 2. Convert text to sequences
x_train_seq = tokenizer.texts_to_sequences(x_train)
x_test_seq = tokenizer.texts_to_sequences(x_test)

# 3. Pad sequences for consistent length
max_length = 100 # Example: Set max length of sequences
x_train_pad = pad_sequences(x_train_seq, maxlen=max_length)
x_test_pad = pad_sequences(x_test_seq, maxlen=max_length)
```

```python
model=Sequential()

model.add(Dense(10,activation='relu'))
model.add(Embedding(10000,64))
model.add(LSTM(64))
model.add(Dense(1,activation='sigmoid'))
model.add(Dropout(0.5))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.fit(x_train_pad, y_train, epochs=10, batch_size=32, validation_data=(x_test_pad, y_test))
```

```python
def recommend_books(genre_input, price_condition='Free', num_recommendations=5):


    genre_encoder = LabelEncoder()
    genre_encoder.fit(book['Genre'])
    genre_encoded = genre_encoder.transform([genre_input])[0]


    filtered_books = book[(book['Genre'] == genre_input)]


    if filtered_books.empty:
        return f"No books found for genre '{genre_input}' that are '{price_condition}'."


    recommendations = filtered_books[['Book', 'Author(s)','Genre']].head(num_recommendations)

    return recommendations
genre_to_search = (input("enter your taste"))  # Example: Looking for fantasy books
recommended_books = recommend_books(genre_to_search, price_condition='Free', num_recommendations=5)
print("Recommended Free Books in the 'Fantasy' Genre:\n", recommended_books)
```
